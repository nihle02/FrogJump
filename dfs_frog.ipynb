```python
start = ['L', 'L', 'L', '_', 'R', 'R', 'R']
goal = ['R', 'R', 'R', '_', 'L', 'L', 'L']

def is_goal(state):
    return state == goal

def get_successors(state):
    successors = []
    blank = state.index('_')
    length = len(state)

    # Kurallar:
    # L sadece sağa zıplar (1 veya 2 taş)
    # R sadece sola zıplar (1 veya 2 taş)

    # L için
    if blank > 0 and state[blank - 1] == 'L':
        new_state = state.copy()
        new_state[blank], new_state[blank - 1] = new_state[blank - 1], '_'
        successors.append(new_state)

    if blank > 1 and state[blank - 2] == 'L' and state[blank - 1] != '_':
        new_state = state.copy()
        new_state[blank], new_state[blank - 2] = new_state[blank - 2], '_'
        successors.append(new_state)

    # R için
    if blank < length - 1 and state[blank + 1] == 'R':
        new_state = state.copy()
        new_state[blank], new_state[blank + 1] = new_state[blank + 1], '_'
        successors.append(new_state)

    if blank < length - 2 and state[blank + 2] == 'R' and state[blank + 1] != '_':
        new_state = state.copy()
        new_state[blank], new_state[blank + 2] = new_state[blank + 2], '_'
        successors.append(new_state)

    return successors

step_counter = 0  # global sayaç

def dfs_verbose(state, visited=None, path=None):
    global step_counter
    if visited is None:
        visited = set()
    if path is None:
        path = []

    state_tuple = tuple(state)
    if state_tuple in visited:
        return None

    visited.add(state_tuple)
    path.append(state)
    print(f"{step_counter}. adım: {state}")
    step_counter += 1

    if is_goal(state):
        print("🎯 Hedefe ulaşıldı!")
        return path

    for successor in get_successors(state):
        result = dfs_verbose(successor, visited.copy(), path.copy())
        if result is not None:
            return result

    # Eğer bu noktaya gelindiyse çıkmaz yola girildi
    print("❌ Çıkmaz yol. Bir önceki adıma geri dönülüyor...")
    if len(path) >= 2:
        back_state = path[-2]
        print(f"{step_counter}. adım (geri dönüş): {back_state}")
        step_counter += 1
    return None


#  fonksiyonu çağır!
dfs_path = dfs_verbose(start)

# dfs_verbose zaten ekrana her adımı yazdığı için tekrar yazmana gerek yok
# sadece varsa final path gösterilebilir (isteğe bağlı)
if dfs_path:
    print("\n📦 Sonuç Yolu:")
    for i, state in enumerate(dfs_path):
        print(f"{i}. adım: {state}")
else:
    print("Çözüm bulunamadı.")


```
