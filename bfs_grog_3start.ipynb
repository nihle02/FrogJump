```python
from collections import deque

def is_goal(state, goal):
    return state == goal

def get_successors(state):
    successors = []
    blank = state.index('_')
    length = len(state)

    # L saÄŸa zÄ±plar
    if blank > 0 and state[blank - 1] == 'L':
        new_state = state.copy()
        new_state[blank], new_state[blank - 1] = new_state[blank - 1], '_'
        successors.append(new_state)

    if blank > 1 and state[blank - 2] == 'L' and state[blank - 1] != '_':
        new_state = state.copy()
        new_state[blank], new_state[blank - 2] = new_state[blank - 2], '_'
        successors.append(new_state)

    # R sola zÄ±plar
    if blank < length - 1 and state[blank + 1] == 'R':
        new_state = state.copy()
        new_state[blank], new_state[blank + 1] = new_state[blank + 1], '_'
        successors.append(new_state)

    if blank < length - 2 and state[blank + 2] == 'R' and state[blank + 1] != '_':
        new_state = state.copy()
        new_state[blank], new_state[blank + 2] = new_state[blank + 2], '_'
        successors.append(new_state)

    return successors

def bfs(start, goal):
    queue = deque()
    queue.append((start, [start]))
    visited = set()

    while queue:
        current, path = queue.popleft()
        state_tuple = tuple(current)

        if state_tuple in visited:
            continue
        visited.add(state_tuple)

        if is_goal(current, goal):
            return path

        for successor in get_successors(current):
            queue.append((successor, path + [successor]))

    return None

def run_scenario(start, goal, scenario_name):
    print(f"\nğŸ”· {scenario_name} BaÅŸlangÄ±Ã§ Durumu: {start}")
    path = bfs(start, goal)
    if path:
        for i, state in enumerate(path):
            print(f"{i}. adÄ±m: {state}")
        print(f"âœ… Toplam adÄ±m sayÄ±sÄ±: {len(path) - 1}")
    else:
        print("ğŸš« Ã‡Ã¶zÃ¼m bulunamadÄ±.")

# SENARYOLAR
start1 = ['L', 'L', 'L', '_', 'R', 'R', 'R']  # klasik dÃ¼zen
start2 = ['L', '_', 'L', 'R', 'L', 'R', 'R']
#start2 = ['L', 'L', '_', 'L', 'R', 'R', 'R']  # boÅŸluk Ã¶nde
start3 = ['L', '_', 'L', 'L', 'R', 'R', 'R']  # boÅŸluk 2. sÄ±rada

goal = ['R', 'R', 'R', '_', 'L', 'L', 'L']

# Ã‡alÄ±ÅŸtÄ±r
run_scenario(start1, goal, "Senaryo 1")
run_scenario(start2, goal, "Senaryo 2")
run_scenario(start3, goal, "Senaryo 3")
```
